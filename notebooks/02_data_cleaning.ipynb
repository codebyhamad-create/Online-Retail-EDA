# 02 - Data Cleaning and Feature Engineering

Cleaning strategy, missing value handling, and creation of analysis-ready features.

```python
import pandas as pd
import numpy as np
from pathlib import Path
import sys
sys.path.append("..")
from src.cleaning import load_raw_data, standardize_columns, clean_retail_data, get_data_quality_summary
from src.feature_engineering import engineer_features, aggregate_customer_metrics
```

## Load and Standardize

```python
DATA_PATH = Path("../data/raw/online_retail_II.csv")
df = load_raw_data(DATA_PATH)
df = standardize_columns(df)
df.head()
```

## Data Quality Assessment

```python
quality = get_data_quality_summary(df)
for k, v in quality.items():
    print(f"{k}: {v}")
```

## Cleaning Strategy

1. Drop rows with missing CustomerID (required for customer analysis)
2. Remove duplicate rows
3. Remove zero or negative prices
4. Keep negative quantities but flag as cancellations (for order status)

```python
df_clean = clean_retail_data(df)
print("Rows before:", len(df))
print("Rows after:", len(df_clean))
print("Rows removed:", len(df) - len(df_clean))
```

## Feature Engineering

```python
df_clean = engineer_features(df_clean)
print("New columns:", [c for c in df_clean.columns if c not in df.columns])
df_clean[["InvoiceDate", "Year", "Month", "DayOfWeek", "Hour", "Sales", "OrderStatus"]].head(10)
```

## Customer Aggregates

```python
customer_metrics = aggregate_customer_metrics(df_clean)
customer_metrics.head(10)
```

## Save Cleaned Data

```python
OUTPUT_PATH = Path("../data/processed/online_retail_cleaned.csv")
df_clean.to_csv(OUTPUT_PATH, index=False)
print(f"Saved to {OUTPUT_PATH}")
```
