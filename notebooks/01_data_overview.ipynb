# 01 - Data Overview

Initial data understanding and quality assessment for the Online Retail II dataset.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/raw/online_retail_II.csv")
df = pd.read_csv(DATA_PATH, encoding="utf-8")
df.head(10)
```

## Data Types and Structure

```python
print("Shape:", df.shape)
print("\nData types:")
print(df.dtypes)
print("\nBasic info:")
df.info()
```

## Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({"Missing": missing, "Percent": missing_pct})
missing_df[missing_df["Missing"] > 0]
```

## Duplicates and Invalid Entries

```python
print("Duplicate rows:", df.duplicated().sum())
print("Negative quantities (refunds/cancellations):", (df["Quantity"] < 0).sum())
print("Zero quantity:", (df["Quantity"] == 0).sum())

price_col = "Price" if "Price" in df.columns else "UnitPrice"
print("Negative price:", (df[price_col] < 0).sum())
print("Zero price:", (df[price_col] == 0).sum())
```

## Country Distribution and Class Imbalance

```python
country_counts = df["Country"].value_counts()
print("Unique countries:", df["Country"].nunique())
print("\nTop 10 countries by transaction count:")
print(country_counts.head(10))
print("\nUK share of total:", (country_counts.get("United Kingdom", 0) / len(df) * 100).round(2), "%")
```

## Pricing and Quantity Summary

```python
price_col = "Price" if "Price" in df.columns else "UnitPrice"
print("Quantity stats:")
print(df["Quantity"].describe())
print("\nPrice stats:")
print(df[price_col].describe())
```

## Data Quality Summary

```python
summary = {
    "Total rows": len(df),
    "Unique invoices": df["Invoice"].nunique(),
    "Unique products": df["StockCode"].nunique(),
    "Unique customers": df["Customer ID"].dropna().nunique() if "Customer ID" in df.columns else "N/A",
    "Date range": f"{df['InvoiceDate'].min()} to {df['InvoiceDate'].max()}",
}
for k, v in summary.items():
    print(f"{k}: {v}")
```
